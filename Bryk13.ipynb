```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro
```

```python
def Box_Muller(size, sd = 1, E = 0):
    U1 = np.random.random(size)
    U2 = np.random.random(size)
    R = np.sqrt(-2 * np.log((1 - U1)))
    Theta = 2 * np.pi * U2
    X = E + sd * R * np.cos(Theta)
    Y = E + sd * R * np.sin(Theta)
    plt.hist(X, density = True)
    plt.show()
    plt.hist(Y, density = True)
    plt.show()
    return X,Y
```

```python
X1, Y1 = Box_Muller(10)
```

```python
X2, Y2 = Box_Muller(100, 100, 2)
```

```python
X3, Y3 = Box_Muller(5000)
```

```python
shapiro(Y3)
```
